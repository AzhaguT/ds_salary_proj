```python
import pandas as pd

df = pd.read_csv('glassdoor_jobs.csv')
```

```python
df = df[df['Salary Estimate'] != '-1'] 
df.head(30)
```

```python
df = df[df['Salary Estimate'] != '-1'] 
df = df[df['Salary Estimate'].notnull()] 
df.info()
```

```python


salary = df['Salary Estimate'].apply(lambda x: x.replace('K','').replace('₹','').replace(',','').split('\n'))
salary = salary.to_frame()
salary = pd.DataFrame(salary['Salary Estimate'].to_list(), columns=['min_salary','max_salary'])
print(type(salary))
df = pd.concat([df,salary], axis=1)
```
